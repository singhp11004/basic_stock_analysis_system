# 🚀 Reinforcement Learning for Stock Market Trading Using PPO

## 1. Introduction

Financial markets are dynamic, stochastic, and highly uncertain. Traditional trading strategies based on fixed rules or classical technical indicators often fail to adapt to rapid market changes.  

Reinforcement Learning (RL) offers a promising alternative, allowing an agent to **learn optimal trading strategies through interaction with the market environment**.  

This project leverages **Proximal Policy Optimization (PPO)**, a state-of-the-art RL algorithm, to develop an agent capable of making sequential trading decisions: **Long, Short, or Hold**, based on historical market data enriched with technical indicators.

---

## 2. Project Objectives

The main goals of this project are:

1. **Design a realistic trading environment**
   - Simulate stock market behavior using historical OHLCV data.
   - Incorporate technical indicators (EMA, RSI, MACD) for richer observations.

2. **Implement a Reinforcement Learning agent using PPO**
   - Train the agent to maximize cumulative profit while managing risk.
   - Explore hyperparameters to optimize learning stability and performance.

3. **Evaluate agent performance on unseen market data**
   - Assess key metrics: final profit, percentage return, Sharpe Ratio, maximum drawdown.
   - Visualize equity curves and drawdowns for performance insight.

4. **Analyze strengths and limitations of RL-based trading**
   - Identify scenarios where RL outperforms traditional strategies.
   - Highlight potential failure modes or overfitting to historical patterns.

---

## 3. Methodology

### 3.1 Environment Design

The trading environment is built with **OpenAI Gym**, offering:

- **Observation Space:** Features include price data (Open, High, Low, Close, Volume) and technical indicators.  
- **Action Space:** Three actions: Hold, Long, Short.  
- **Reward Function:** Profit/loss from trades, adjusted for transaction costs and risk penalties based on volatility.  
- **Episode Termination:** Ends when historical data is exhausted.

This setup provides a realistic yet computationally efficient simulation of trading.

### 3.2 Proximal Policy Optimization (PPO)

PPO is chosen for its **learning stability and sample efficiency**. Key aspects:

- **Policy Gradient Optimization:** Agent learns a stochastic policy mapping states to actions.  
- **Clipped Objective Function:** Prevents drastic policy updates that could degrade performance.  
- **Generalized Advantage Estimation (GAE):** Reduces variance in updates while maintaining low bias.

Training involves multiple episodes where the agent interacts with the environment, receives rewards, and iteratively improves its trading policy.

## 4. Data and Feature Engineering

Historical stock data is obtained from Yahoo Finance (OHLCV). Preprocessing includes:

- **Cleaning:** Removing missing or inconsistent entries.  
- **Technical Indicators:**  
  - **EMA (Exponential Moving Average):** Trend direction.  
  - **RSI (Relative Strength Index):** Momentum and overbought/oversold conditions.  
  - **MACD (Moving Average Convergence Divergence):** Trend reversal signals.  

These features provide a richer observation space for the RL agent.

## 5. Evaluation Metrics

Performance is assessed using:

- **Final Profit:** Net gain at the end of the test period.  
- **Percentage Return:** Relative gain as a percentage of initial capital.  
- **Sharpe Ratio:** Risk-adjusted return.  
- **Maximum Drawdown:** Largest peak-to-trough decline, representing downside risk.

Equity curves and drawdown charts visually complement the metrics.

```python
# ============================================================
# 📦 Required Libraries for RL Trading Project
# ============================================================

# -------------------------------
# Data Manipulation and Analysis
# -------------------------------
import pandas as pd      # DataFrames, CSV read/write, basic analysis
import numpy as np       # Numerical computations, arrays

# -------------------------------
# Visualization
# -------------------------------
import matplotlib.pyplot as plt   # Plotting equity curves, drawdowns, etc.

# -------------------------------
# Financial Data and Technical Indicators
# -------------------------------
import yfinance as yf   # Download historical market data
import ta               # Compute technical indicators (EMA, RSI, MACD, etc.)

# -------------------------------
# Reinforcement Learning
# -------------------------------
import gym                               # Base class for custom environments
from gym import spaces                     # Define action and observation spaces
from stable_baselines3 import PPO         # Proximal Policy Optimization algorithm
from stable_baselines3.common.vec_env import DummyVecEnv  # Vectorized environment wrapper
```

```python
# ============================================================
# 📊 Stock Data Loader using Yahoo Finance API
# ------------------------------------------------------------
# This script fetches real-world financial market data for a
# specified stock symbol (e.g., AAPL) using Yahoo Finance.
# It then returns a cleaned pandas DataFrame containing
# key columns such as Open, High, Low, Close, and Volume.
# ============================================================


# ============================================================
# ⚙️ Function: load_real_data
# ------------------------------------------------------------
# Purpose:
#   Downloads historical stock data for a given symbol and time range
#   from Yahoo Finance, cleans it, and returns a structured DataFrame.
#
# Parameters:
#   symbol (str): The stock ticker symbol to download (default = 'AAPL').
#   start (str):  The start date for historical data in 'YYYY-MM-DD' format.
#   end (str):    The end date for historical data in 'YYYY-MM-DD' format.
#
# Returns:
#   pd.DataFrame: A DataFrame containing OHLCV (Open, High, Low, Close, Volume) data.
# ============================================================
def load_real_data(symbol='AAPL', start='2020-01-01', end='2024-01-01'):
    # 🧭 Download real market data from Yahoo Finance
    df = yf.download(symbol, start=start, end=end)
    
    # 🧹 Keep only essential columns and remove missing values
    df = df[['Open', 'High', 'Low', 'Close', 'Volume']].dropna()
    
    # 📤 Return the cleaned DataFrame
    return df


# ============================================================
# 🚀 Example Execution
# ------------------------------------------------------------
# Demonstration: Load Apple (AAPL) stock data from 2020 to 2024
# ============================================================
df = load_real_data('AAPL')
df
```

```python
# ============================================================
# 📈 Technical Indicators Calculation using 'ta' Library
# ------------------------------------------------------------
# This script calculates common technical indicators (EMA, RSI,
# and MACD) using the 'ta' library for financial time series data.
# The indicators are added as new columns to an existing DataFrame.
# ============================================================



# ============================================================
# 🧭 Preprocessing
# ------------------------------------------------------------
# Assumes that a DataFrame named `df` already exists and contains
# at least the 'Close' column representing closing prices.
# This section ensures the 'Close' data is a 1D pandas Series.
# ============================================================

close_series = df["Close"]

# ✅ Ensure that 'close_series' is a 1D Series, not a DataFrame
if isinstance(close_series, pd.DataFrame):
    close_series = close_series.iloc[:, 0]

# ============================================================
# ⚙️ Technical Indicator Calculations
# ------------------------------------------------------------
# Below we compute three popular indicators:
#   1. EMA (Exponential Moving Average)
#   2. RSI (Relative Strength Index)
#   3. MACD (Moving Average Convergence Divergence)
# Each of these is appended to the DataFrame as a new column.
# ============================================================

# 1️⃣ Exponential Moving Average (EMA) — 20 periods
df["EMA_20"] = ta.trend.ema_indicator(close=close_series, window=20)

# 2️⃣ Relative Strength Index (RSI) — 14 periods
df["RSI_14"] = ta.momentum.rsi(close=close_series, window=14)

# 3️⃣ Moving Average Convergence Divergence (MACD)
macd = ta.trend.MACD(close=close_series)
df["MACD"] = macd.macd()                 # MACD line
df["MACD_Signal"] = macd.macd_signal()   # Signal line
df["MACD_Hist"] = macd.macd_diff()       # Histogram (difference)

# ============================================================
# 🧹 Post-processing
# ------------------------------------------------------------
# Replace any NaN values (caused by indicator warm-up periods)
# with zeros for cleaner visualization or further processing.
# ============================================================
df.fillna(0, inplace=True)

# ============================================================
# 👀 Display Result
# ------------------------------------------------------------
# Show the last few rows of the DataFrame with added indicators
# ============================================================
df
```

```python
# ============================================================
# 🤖 Custom Trading Environment for Reinforcement Learning
# ------------------------------------------------------------
# This environment simulates a simple stock trading scenario
# for a reinforcement learning agent (Long, Short, Hold).
# It supports step-wise reward calculation, balance tracking,
# and observation of technical indicators and prices.
# ============================================================

 

# ============================================================
# ⚙️ TradingEnv Class Definition
# ============================================================
class TradingEnv(gym.Env):
    """
    A simple trading environment using real market data for RL.

    Actions:
        0 - Hold
        1 - Long (buy)
        2 - Short (sell)

    Observations:
        Feature vector including prices and technical indicators.

    Rewards:
        Based on profit/loss from trades, updated per step.
    """
    metadata = {'render.modes': ['human']}

    # ===============================
    # 🔹 Initialization
    # ===============================
    def __init__(self, df, initial_balance=10000):
        """
        Args:
            df (pd.DataFrame): Historical market data with 'Close' column
            initial_balance (float): Starting cash balance
        """
        super(TradingEnv, self).__init__()

        self.df = df.reset_index(drop=True)
        self.initial_balance = initial_balance

        # Observation space: features (OHLCV + indicators)
        self.observation_space = spaces.Box(
            low=-np.inf, high=np.inf,
            shape=(self.df.shape[1],), dtype=np.float32
        )

        # Action space: 0=Hold, 1=Long, 2=Short
        self.action_space = spaces.Discrete(3)

        # Initialize environment state
        self.reset()

    # ===============================
    # 🔹 Reset Environment
    # ===============================
    def reset(self):
        """
        Reset environment to initial state at episode start.

        Returns:
            np.array: initial observation
        """
        self.balance = self.initial_balance
        self.position = 0      # 1=Long, -1=Short, 0=None
        self.entry_price = 0   # Entry price for current position
        self.current_step = 0
        self.done = False

        return self._get_observation()

    # ===============================
    # 🔹 Get Current Observation
    # ===============================
    def _get_observation(self):
        """
        Return current step data as agent observation.

        Returns:
            np.array: observation vector
        """
        obs = self.df.iloc[self.current_step].values.astype(np.float32)
        return obs

    # ===============================
    # 🔹 Step Function
    # ===============================
    def step(self, action):
        """
        Execute one environment step given the agent's action.

        Args:
            action (int): 0=Hold, 1=Long, 2=Short

        Returns:
            obs (np.array): next observation
            reward (float): step reward
            done (bool): episode finished
            info (dict): additional info (balance, position)
        """
        if self.done:
            raise RuntimeError("Environment needs reset before reuse.")

        # Get current and next prices
        current_price = self.df.loc[self.current_step, "Close"]
        next_price = self.df.loc[min(self.current_step + 1, len(self.df)-1), "Close"]
        reward = 0.0

        # -------------------------------
        # 1️⃣ Basic Trading Logic
        # -------------------------------
        if action == 1:  # Long
            if self.position == 0:
                self.position = 1
                self.entry_price = current_price
            elif self.position == -1:  # Close Short, open Long
                reward = self.entry_price - current_price
                self.balance += reward
                self.position = 1
                self.entry_price = current_price

        elif action == 2:  # Short
            if self.position == 0:
                self.position = -1
                self.entry_price = current_price
            elif self.position == 1:  # Close Long, open Short
                reward = current_price - self.entry_price
                self.balance += reward
                self.position = -1
                self.entry_price = current_price

        # -------------------------------
        # 2️⃣ Reward for Holding Position
        # -------------------------------
        if self.position == 1:       # Long
            reward = next_price - current_price
        elif self.position == -1:    # Short
            reward = current_price - next_price

        # Ensure reward and balance are scalar
        if isinstance(reward, pd.Series):
            reward = reward.iloc[0]
        if isinstance(self.balance, pd.Series):
            self.balance = self.balance.iloc[0]

        # Update balance and step
        self.balance += reward
        self.current_step += 1

        # Check if episode done
        done = self.current_step >= len(self.df) - 2
        self.done = done

        info = {
            "balance": self.balance,
            "position": self.position,
        }

        obs = self._get_observation()
        return obs, reward, done, info

    # ===============================
    # 🔹 Render Environment
    # ===============================
    def render(self, mode='human'):
        """
        Print current environment status.
        """
        balance_scalar = self.balance.iloc[0] if isinstance(self.balance, pd.Series) else self.balance
        position_scalar = self.position.iloc[0] if isinstance(self.position, pd.Series) else self.position

        print(f"Step: {self.current_step} | Balance: {balance_scalar:.2f} | Position: {position_scalar}")
```

```python
# ============================================================
# 🏃 Running the Trading Environment (Sample Episode)
# ------------------------------------------------------------
# This script demonstrates how to initialize and run the
# custom TradingEnv for a few steps with random actions.
# ============================================================

# ===============================
# 🔹 Initialize Environment
# ===============================
# Use the previously prepared DataFrame `df` with OHLCV + indicators
env = TradingEnv(df)

# ===============================
# 🔹 Reset Environment
# ===============================
# Returns the initial observation
obs = env.reset()

# ===============================
# 🔹 Run a Short Simulation
# ------------------------------
# Take 5 random actions to demonstrate environment dynamics
# ===============================
for step in range(5):
    # Randomly sample an action from action space: Hold, Long, Short
    action = env.action_space.sample()

    # Execute the action in the environment
    obs, reward, done, info = env.step(action)

    # Render the environment state to console
    env.render()

    # Optional: break if episode is done
    if done:
        print("Episode finished early.")
        break
```

```python
# ============================================================
# 🚀 Training a Reinforcement Learning Agent (PPO) on Real Market Data
# ------------------------------------------------------------
# This script demonstrates how to train a PPO agent from
# Stable-Baselines3 on a custom TradingEnv using real market data.
# ============================================================


# ===============================
# 1️⃣ Prepare the Trading Environment
# -------------------------------
# Wrap the custom TradingEnv inside DummyVecEnv for compatibility
# with Stable-Baselines3 algorithms
# ===============================
env = DummyVecEnv([lambda: TradingEnv(df)])

# ===============================
# 2️⃣ Create PPO Model
# -------------------------------
# Policy: MlpPolicy (simple multi-layer perceptron)
# hyperparameters:
#   - learning_rate: 0.0003
#   - n_steps: 128 (steps per update)
#   - batch_size: 64
#   - gamma: 0.99 (discount factor)
#   - gae_lambda: 0.95 (GAE factor)
# ===============================
model = PPO(
    policy="MlpPolicy",
    env=env,
    verbose=1,
    learning_rate=0.0003,
    n_steps=128,
    batch_size=64,
    gamma=0.99,
    gae_lambda=0.95,
)

# ===============================
# 3️⃣ Train the PPO Agent
# -------------------------------
# total_timesteps: number of environment interactions
# Adjust this value depending on dataset size and training budget
# ===============================
print("🚀 Starting RL agent training on real market data...")
model.learn(total_timesteps=100_000)

# ===============================
# 4️⃣ Save Trained Model
# -------------------------------
# The trained PPO agent is saved to disk for later evaluation
# or deployment.
# ===============================
model.save("ppo_trading_real")
print("✅ Model saved at: ppo_trading_real.zip")
```

```python
# ============================================================
# 📊 Evaluate Trained PPO Agent on Real Market Data
# ------------------------------------------------------------
# This script runs a trained PPO agent on a TradingEnv using
# real market data and records performance metrics.
# It also plots the equity curve and saves the results to CSV.
# ============================================================



# ===============================
# 1️⃣ Prepare Evaluation Environment
# -------------------------------
# Create a fresh TradingEnv for evaluation (no vectorization)
# and reset to initial state
# ===============================
env = TradingEnv(df)
obs = env.reset()

# Lists to store evaluation metrics
balances = []   # Track agent's cash balance over time
positions = []  # Track agent's position (Hold/Long/Short)
rewards = []    # Track reward at each step

print("🚀 Starting evaluation of the RL agent on real market data...")

# ===============================
# 2️⃣ Run the Agent on the Entire Dataset
# -------------------------------
done = False
step = 0

while not done:
    # Predict the action using the trained PPO model
    action, _ = model.predict(obs)

    # Take a step in the environment with the predicted action
    obs, reward, done, info = env.step(int(action))

    # Record metrics
    balances.append(info["balance"])
    positions.append(info["position"])
    rewards.append(reward)

    step += 1

    # Periodically print progress
    if step % 100 == 0:
        print(f"Step: {step} | Balance: {info['balance']:.2f} | Reward: {reward:.4f}")

# ===============================
# 3️⃣ Plot Equity Curve
# -------------------------------
# Visualize the agent's balance over time
# ===============================
plt.figure(figsize=(10, 5))
plt.plot(balances, label='Equity Curve', color='green')
plt.title('RL Agent Performance on Real Market Data')
plt.xlabel('Time Step')
plt.ylabel('Balance / Equity')
plt.legend()
plt.grid(True)
plt.show()

# ===============================
# 4️⃣ Save Evaluation Results
# -------------------------------
# Store step-wise metrics in a CSV file for further analysis
# ===============================
results = pd.DataFrame({
    "Step": range(len(balances)),
    "Balance": balances,
    "Reward": rewards,
    "Position": positions
})

results.to_csv("evaluation_results.csv", index=False)
print("✅ Evaluation results saved at: evaluation_results.csv")
```

```python
# ============================================================
# 📊 RL Agent Performance Analysis
# ------------------------------------------------------------
# This script analyzes the evaluation results of a trained
# RL agent, computing key performance metrics such as:
#   - Final Profit
#   - Percentage Return
#   - Sharpe Ratio
#   - Maximum Drawdown
# It also visualizes the equity curve along with drawdowns.
# ============================================================

# ===============================
# 1️⃣ Load Evaluation Results
# -------------------------------
# CSV contains step-wise metrics from agent evaluation
# Columns: Step, Balance, Reward, Position
# ===============================
df = pd.read_csv("evaluation_results.csv")

# ===============================
# 2️⃣ Compute Performance Metrics
# -------------------------------

# 💰 Final Profit (difference between last and first balance)
final_profit = df['Balance'].iloc[-1] - df['Balance'].iloc[0]

# 📈 Percentage Return
return_percentage = (final_profit / df['Balance'].iloc[0]) * 100

# 🔹 Daily changes in balance (step-wise profit/loss)
df['daily_change'] = df['Balance'].diff()

# ⚙️ Sharpe Ratio (return-to-risk ratio)
# Note: small epsilon added to avoid division by zero
sharpe_ratio = np.mean(df['daily_change']) / (np.std(df['daily_change']) + 1e-9) * np.sqrt(len(df))

# 📉 Maximum Drawdown
rolling_max = df['Balance'].cummax()      # Running maximum balance
drawdown = df['Balance'] / rolling_max - 1.0
max_drawdown = drawdown.min()

# ===============================
# 3️⃣ Print Metrics
# -------------------------------
print(f"💰 Final Profit: {final_profit:.2f}")
print(f"📈 Return Percentage: {return_percentage:.2f}%")
print(f"⚙️ Sharpe Ratio: {sharpe_ratio:.3f}")
print(f"📉 Max Drawdown: {max_drawdown:.3f}")

# ===============================
# 4️⃣ Plot Equity Curve and Drawdown
# -------------------------------
fig, ax1 = plt.subplots(figsize=(10, 5))

# Equity Curve
ax1.plot(df['Balance'], color='green', label='Equity')
ax1.set_xlabel('Steps')
ax1.set_ylabel('Equity', color='green')

# Drawdown (secondary axis)
ax2 = ax1.twinx()
ax2.plot(drawdown, color='red', alpha=0.4, label='Drawdown')
ax2.set_ylabel('Drawdown', color='red')

# Title, Legend, Grid
plt.title("RL Agent Performance Analysis (Equity & Drawdown)")
fig.legend(loc="upper left")
plt.grid(True)
plt.show()
```
